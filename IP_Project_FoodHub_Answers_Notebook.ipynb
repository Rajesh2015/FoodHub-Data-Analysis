Description
Context
The number of restaurants in New York is increasing day by day. Lots of students and busy professionals rely on those restaurants due to their hectic lifestyles. Online food delivery service is a great option for them. It provides them with good food from their favorite restaurants. A food aggregator company FoodHub offers access to multiple restaurants through a single smartphone app.

The app allows restaurants to receive a direct online order from a customer. The app assigns a delivery person from the company to pick up the order after it is confirmed by the restaurant. The delivery person then uses the map to reach the restaurant and waits for the food package. Once the food package is handed over to the delivery person, he/she confirms the pick-up in the app and travels to the customer's location to deliver the food. The delivery person confirms the drop-off in the app after delivering the food package to the customer. The customer can rate the order in the app. The food aggregator earns money by collecting a fixed margin of the delivery order from the restaurants.

Objective
The food aggregator company has stored the data of the different orders made by the registered customers in their online portal. They want to analyze the data to get a fair idea about the demand of different restaurants which will help them in enhancing their customer experience. Suppose you are hired as a Data Scientist in this company and the Data Science team has shared some of the key questions that need to be answered. ## Data Description
The data contains the different data related to a food order. The detailed data dictionary is given below.

### Data Dictionary:

* order_id: Unique ID of the order

**Utility Function**

```python
def find_top_or_bottom(df, col_name, n, isTop):
    """
    Find the top or bottom 'n' occurrences in a specified column of a DataFrame.

    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame to be analyzed.
    col_name : str
        The name of the column to analyze.
    n : int
        The number of top or bottom occurrences to find.
    isTop : bool
        If True, find the top 'n' occurrences. If False, find the bottom 'n' occurrences.

    Returns:
    --------
    pandas.DataFrame
        A DataFrame containing only the rows with the top or bottom 'n' occurrences in the specified column.
    """
    if isTop:
        top_values = df[col_name].value_counts().nlargest(n).index
        top_df = df[df[col_name].isin(top_values)]
        return top_df
    else:
        bottom_values = df[col_name].value_counts().nsmallest(n).index
        bottom_df = df[df[col_name].isin(bottom_values)]
        return bottom_df
```

```python
def anotate_chart(plot, ax):
    """
    Annotates the bars of a bar chart with their height values.

    Parameters:
    plot : matplotlib.pyplot
        The plot object to be annotated.
    ax : matplotlib.axes._axes.Axes
        The axes object containing the bar chart.

    Description:
    This function iterates through the patches (bars) in the axes object and annotates each bar with its height.
    The annotations are centered horizontally on the bar and placed slightly above the top of the bar.
    """

    for p in ax.patches:
        plot.annotate(format(p.get_height(), '.0f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center',
                      xytext=(0, 9),
                      textcoords='offset points')


```

```python
def draw_count_plot(df, x, n, xlabel, ylabel, title, isTop, hue=None):
    """
    Draws a count plot for the top or bottom `n` categories in a DataFrame and annotates it with counts.

    Parameters:
    df : pandas.DataFrame
        The DataFrame containing the data to be plotted.
    x : str
        The column name to be used for the x-axis in the count plot.
    n : int
        The number of top or bottom categories to be plotted.
    xlabel : str
        The label for the x-axis.
    ylabel : str
        The label for the y-axis.
    title : str
        The title of the plot.
    isTop : bool
        If True, the plot will show the top `n` categories. If False, the plot will show the bottom `n` categories.
    hue : str, optional
        The column name for the hue parameter, which adds a third dimension by splitting bars in the count plot.

    Returns:
    matplotlib.pyplot
        The plot object containing the count plot with annotations.
    """
    top_resturant_df = find_top_or_bottom(df, x, n, isTop)
    plt.figure(figsize=(12, 6))
    ax = sns.countplot(data=top_resturant_df, x=x, hue=hue, color='red')
    plt.xticks(rotation=45)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    anotate_chart(plt, ax)
    return plt
```

```python
def draw_box_plot_with_legend(df, x, xlabel):
    """
    Draw a box plot for a specified column in a DataFrame with a legend indicating the mean value.

    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame to be analyzed.
    x : str
        The name of the column to be plotted.
    xlabel : str
        The label for the x-axis.

    Returns:
    --------
    matplotlib.pyplot
        The plot object.
    """
    mean_value = df[x].mean()
    sns.boxplot(data=df, x=x, color='red')
    plt.xlabel(xlabel)
    plt.axvline(mean_value, color='green', linestyle=':', label=f'Mean: {mean_value:.2f}')
    plt.legend()
    return plt
```

```python
def calculatePercentage(df):
  """
    Calculates the percentage distribution of customer frequencies in a DataFrame.

    Parameters:
    df : pandas.DataFrame
        The DataFrame containing customer data. It must include a column named 'customer_id'.

    Returns:
    pandas.Series
        A Series with the percentage distribution of the frequencies of customer occurrences.
  """
  customer_count_df = df['customer_id'].value_counts().reset_index()
  frequency_count= customer_count_df['count'].value_counts()

  finalres=frequency_count['Percentage']=frequency_count/customer_count_df.shape[0]*100
  return finalres
```

**Utility**

```python
def plot_bar_distribution(data, x, y, hue, title, x_label, y_label, hue_title, color, figsize=(12, 8), rotation=90):
    """
    Plots a bar plot showing the distribution of a categorical variable across different bins.

    Parameters:
    - data: DataFrame containing the data to plot.
    - x: Column name for the x-axis (categorical variable).
    - y: Column name for the y-axis (percentage or count).
    - hue: Column name for the hue (bin categories).
    - title: Title of the plot.
    - x_label: Label for the x-axis.
    - y_label: Label for the y-axis.
    - hue_title: Title for the hue legend.
    - color: Color of the bars.
    - figsize: Size of the figure.
    - rotation: Rotation angle for x-axis labels.
    """
    plt.figure(figsize=figsize)
    sns.barplot(data=data, x=x, y=y, hue=hue, palette='Reds')
    plt.xticks(rotation=rotation)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend(title=hue_title)
    return plt
```

### **Question 14:** The company charges the restaurant 25% on the orders having cost greater than 20 dollars and 15% on the orders having cost greater than 5 dollars. Find the net revenue generated by the company across all orders. [3 marks]

```python
def calculate_profit(cost):
    """
    Calculate total revenue depending on given conditions
    Parameters:
    - cost: cost_of_order
    """
    if cost > 20:
        return cost * 0.25
    elif (cost > 5) & (cost<=20):
        return cost * 0.15
    else:
        return 0
```

```python
food_df['revenue']=food_df['cost_of_the_order'].apply(calculate_profit)
print(f"Total Revenue is {food_df['revenue'].sum()}")
```

#### Observations:
Total revenue on all the orders is around=6166.303
